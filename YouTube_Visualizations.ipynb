```python
import numpy as np # arrays
import pandas as pd # dataframes
import matplotlib.pyplot as plt # plotting
import seaborn as sns # prettier plotting
```

```python
youtube = pd.read_csv('youtube.csv')
```

```python
youtube = youtube.rename(columns={"Subscribers(millions)":"Subscribers"})
```

```python
youtube.groupby('Language')['Subscribers'].mean().reset_index()
```

```python
categories = youtube.groupby('Category')['Channel'].count().reset_index()
categories
```

```python
countries = youtube.groupby('Country')['Channel'].count().reset_index()
countries
```

```python
languages = youtube.groupby('Language')['Channel'].count().reset_index()
languages
```

```python
plt.figure(dpi=125)
p1 = plt.hist(youtube['Subscribers'])
plt.xlabel('Subscribers in Millions')
plt.ylabel('Y')
plt.title('X')
```

```python
plt.figure(dpi=150, figsize=(10,5))
plt.barh(countries['Country'], countries['Channel'], color="green")
```

```python
countries = countries.sort_values('Channel', ascending=False)
countries
```

```python
plt.figure(dpi=200)
sns.barplot(data=countries, y='Country', x='Channel', palette='Reds_r')
plt.xlabel('Count within Top 50')
plt.title('Top Countries among Top 50 YouTube Channels')
```

```python
categories = categories.sort_values('Channel', ascending=False)

plt.figure(dpi=300, figsize = (8,4))
sns.barplot(data=categories, y='Category', x='Channel')
plt.xlabel('Count within Top 50')
plt.ylabel('Channel Type')
plt.title('Common Channel Types Among Top 50 YouTube Channels')
```
